```python
import matplotlib.pyplot as plt
import matplotlib.figure as md
import numpy as np

%matplotlib notebook

a = 0
b = 3
c = 2

n = 1000  #Numero de Particiones

h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = (8 * x)-16  #Función a Integrar corresponde a Fuerza F(x)

I_simp_F = (h/3) * (f[a] + 2*sum(f[:n-2:2]) + 4*sum(f[1:n-1:2]) + f[b]) # Integración por Simpson
err_simp_F = -12 - I_simp_F  #Valor Real de la Integración es (-12J)
print("Aproximación Integral por Simpson de F(x)=8x-16 :", I_simp_F)
print("Error Aproximación Integral por Simpson : ",err_simp_F)
plt.figure()
plt.plot(x,f)
plt.fill_between(x,f, color = 'orange', alpha = 0.5)
plt.show()


# Energia Potencial U(x)= Integral F(x)dx en [0,2] 
y = np.linspace(a, c, n)
r = (8 * y)-16  #Función a Integrar corresponde a Fuerza F(x)
g = (c - a) / (n - 1)
I_simp_U = (g/3) * (f[a] + 2*sum(r[:n-2:2]) + 4*sum(r[1:n-1:2]) + r[c]) # Integración por Simpson
print("Energíal Potencia U(x) [0-2]:", I_simp_U)
plt.figure()
plt.plot(y,r)
plt.fill_between(y,r, color = 'red', alpha = 0.5)
plt.show()
```
