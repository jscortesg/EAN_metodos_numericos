```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
t0 = 0
tf = 3

N1 = 81

h = (tf-t0)/(N1-1)

intt1 = np.linspace(t0, tf, N1)

F = 8*(intt1) - 16

Deltax_t = round(0.5*h*(sum(F[0:N1-1]) + sum(F[1:N1-1])), 2)
```

```python
Deltax_t
```

```python
t0 = 0
tf = 2

N1 = 81

hs1 = (tf-t0)/(N1-1)

intt2 = np.linspace(t0, tf, N1)

F1 = 8*(intt2) - 16

Deltax_t1 = round(0.5*hs1*(sum(F1[0:N1-1]) + sum(F1[1:N1-1])),2)

plt.figure()

plt.plot(intt2, F1)

plt.fill_between(intt2, F1, color = 'orange', alpha = 0.5)

plt.show()
```
