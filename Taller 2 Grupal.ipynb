```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
##EJERCICIO 1 CLASE
x0 = 0
xf = 3

N1 = 10

hs1 = (xf-x0)/(N1-1)

intt1 = np.linspace(x0, xf, N1)

w = 8*intt1 -16

Deltax_t = round(0.5*hs1*(sum(w[0:N1-1]) + sum(w[1:N1-1])),2)
```

```python
print(Deltax_t)
```

```python
##EJERCICIO 2 CLASE
x0 = 0
xf = 2

N1 = 10

hs1 = (xf-x0)/(N1-1)

intt2 = np.linspace(x0, xf, N1)

w = -(8*intt2 -16)

Deltax_2 = round(0.5*hs1*(sum(w[0:N1-1]) + sum(w[1:N1-1])),2)
```

```python
print(Deltax_2)
```

```python

x= np.linspace (0,2,100)
y= -4*pow(x,2)+16*x
plt.plot(x,y,label = 'U(x)= -4 x^2+16x')
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Grafico de U(x)= -4 x^2+16x')
plt.legend
plt.grid(True)
plt.show()
```
