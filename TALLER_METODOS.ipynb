```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
    return 8 * x - 16
def int_t(f, a, b, n):
    h = (b - a) / n
    suma = 0.5 * (f(a) + f(b))

    for i in range(1, n):
        suma += f(a + i * h)

    return h * suma
x_0 = 0
x_f = 3

intervalos = 1000

res_int = int_t(F, x_0, x_f, intervalos)
print("la fuerza realizada:", res_int, "J")
```

```python
def force(x):
    return 8*x - 16
```

```python
def int_trape(func, a, b, n):
    x= np.linspace(a, b, n+1)
    y= func(x)
    h = (b - a) / n
    inte = h * (0.5*y[0] + sum(y[1:-1]) + 0.5*y[-1])
    return inte
a, b = 0, 2
n = 100
x = np.linspace(a, b, n)
u = [1]
```

```python
for i in range(1, n):
    integral = int_trape(force, a, x[i], n)
    u.append(u[-1] + integral)
```

```python
plt.plot(x, u, label='U(x)')
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Energ√≠a Potencial U(x) - Regla del Trapecio')
plt.legend()
plt.grid(True)
plt.show()
```
