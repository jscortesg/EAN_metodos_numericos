```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

a = 0
b = 3
n = 1000
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)

f = 8*x - 16
#def f(x):
#    return 8*x - 16

#I_simp = (h/3) * (f(x[0]) + 2*sum(f(x[:n-2:2])) + 4*sum(f(x[1:n-1:2])) + f(x[n-1]))

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) + 4*sum(f[1:n-1:2]) + f[n-1]) 
err_sim = - 12 - I_simp

plt.figure()
plt.plot(x,f)
plt.fill_between(x,f, color = 'orange', alpha = 0.5)
plt.show()
print(I_simp)
print(err_sim)
```

grafica integracion Numerica

```python
fig, ax = plt.subplots(figsize=(8,8))


a = 0
b = 2
N = 50
h = (b - a)/(N-1)

x = np.linspace(a, b, N)

def fuerza(x):
    return 8 * x - 16

def integral(x):
    # integración numérica proporcionada por scipy
    result, _ = odeint(lambda u, x: -fuerza(x), 1, [0, x], full_output=True)
    return result[-1]

U = [integral(i) for i in x]

ax.plot(x, U, c='red', alpha=0.5, drawstyle="steps", label='Integracion Numerica')
ax.plot(x, U, c='black', label='función')
leg = ax.legend(prop={"size":20})
```
