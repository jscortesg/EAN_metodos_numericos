# Punto A

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
    return 8*x - 16

x0 = 0
xf = 3 
```

```python
def calcular_trabajo(F, x0, xf, n):
    x = np.linspace(x0, xf, n)
    y = F(x)
    dx = (xf - x0) / (n - 1)
    
   
    trabajo = 0.5 * (y[0] + y[-1]) + np.sum(y[1:-1])
    trabajo *= dx
    
    return trabajo
```

```python
total = calcular_trabajo(F, x0, xf, n=10)
```

# PUNTO B

```python
def U(x):
    return -4 * x**2 + 16 * x + 1
```

```python
valordex = np.linspace(0, 2, 100)
valordeu = U(valordex)
```

```python
# Gr√°fico
plt.plot(valordex, valordeu, label='U(x) = -4x^2 + 16x + 1')
plt.xlabel('x (m)')
plt.ylabel('U(x) (J)')
plt.legend()
plt.show()
```
