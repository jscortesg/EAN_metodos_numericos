```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F = lambda x: 8*x - 16 # Fuerza
x0 = 0 
xf = 3
dx = 0.1
```

```python
trabajo = 0
for i in range(int((xf-x0)/dx)):
    x1 = x0 + i*dx
    x2 = x0 + (i+1)*dx
    trabajo += dx*(F(x1) + F(x2))/2
print('Trabajo =',trabajo)
```

```python
U = [1] 
x = [0]
for i in range(int((2-x0)/dx)):
    x1 = x[-1]  
    x2 = x1 + dx
    U.append(U[-1] - dx*(F(x1) + F(x2))/2)
    x.append(x2)
```

```python
plt.plot(x,U)
plt.xlabel('x (m)')
plt.ylabel('U(x) (J)')
plt.title('EnergÃ­a potencial')
plt.show()
```
