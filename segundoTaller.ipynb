b) Recordando que $ F(x) = - \frac{dU}{dx} $ y tomando $U(0) = 1 J$, encuentre la gráfica de la energía potencial asociada $U(x)$ en el intervalo $ 0\leq x \leq 2 $

```python
def f(x):
    return 8*x - 16

def energia(a, b, n):
    h = (b - a) / (n-1)
    resultado = 0.5 * (f(a) + f(b))

    for i in range(1, n):
        resultado += f(a + i * h)

    return -h * resultado 

a = 0
b = 2
n = 81

u0 = 1
potencia = [u0]


for i in range(1, n + 1):
    x = a + i * (b - a) / (n-1)
    valor_energia = energia(a, x, n)
    potencia.append(valor_energia + u0)

x = [a + i * (b - a) / (n-1) for i in range(n + 1)]

plt.plot(x, potencia)
plt.xlabel('x')
plt.ylabel('U(x)')
plt.show()
```
